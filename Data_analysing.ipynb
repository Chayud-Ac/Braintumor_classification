```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np


train_dir = 'Training'
test_dir = 'Testing'


def get_image_paths(directory):
    classes = os.listdir(directory)
    image_paths = []
    labels = []
    for label in classes:
        class_path = os.path.join(directory, label)
        if os.path.isdir(class_path):
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                if os.path.isfile(img_path):
                    image_paths.append(img_path)
                    labels.append(label)
    return image_paths, labels

def plot_label_distribution(labels, title):
    plt.figure(figsize=(10, 5))
    plt.hist(labels, bins=len(set(labels)), edgecolor='black')
    plt.title(title)
    plt.xlabel('Class')
    plt.ylabel('Number of images')
    plt.xticks(rotation=45)
    plt.show()



train_image_paths, train_labels = get_image_paths(train_dir)
test_image_paths, test_labels = get_image_paths(test_dir)

plot_label_distribution(train_labels, 'Training Set Label Distribution')
plot_label_distribution(test_labels, 'Testing Set Label Distribution')

```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def plot_random_images(image_paths, labels, num_images=10, title='Random Image Samples'):
    selected_indices = random.sample(range(len(image_paths)), num_images)
    selected_paths = [image_paths[i] for i in selected_indices]
    selected_labels = [labels[i] for i in selected_indices]
    
    plt.figure(figsize=(15, 10))
    for i, img_path in enumerate(selected_paths):
        img = Image.open(img_path)
        plt.subplot(2, num_images // 2, i + 1)
        plt.imshow(img)
        plt.title(selected_labels[i])
        plt.axis('off')
    plt.suptitle(title)
    plt.show()


# Plot random image samples
plot_random_images(train_image_paths, train_labels, num_images=10, title='Random Training Image Samples')
plot_random_images(test_image_paths, test_labels, num_images=10, title='Random Testing Image Samples')
```
